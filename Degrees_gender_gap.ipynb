```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
cb_light_grey = (171/255, 171/255, 171/255)
stem_cats = ['Psychology','Biology', 'Math and Statistics','Physical Sciences','Computer Science',  'Engineering' ]
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

majors = [stem_cats,lib_arts_cats,other_cats]

fig = plt.figure(figsize=(18, 18))

for sp in range(0,3):
    for m in range(len(majors[sp])):
        plot_no = m*3+sp+1
        ax = fig.add_subplot(6,3,plot_no)
    
        ax.plot(women_degrees['Year'], women_degrees[majors[sp][m]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[majors[sp][m]], c=cb_orange, label='Men', linewidth=3)
        ax.spines["right"].set_visible(False)    
        ax.spines["left"].set_visible(False)
        ax.spines["top"].set_visible(False)    
        ax.spines["bottom"].set_visible(False)
        ax.set_xlim(1968, 2011)
        ax.set_ylim(0,100)
        ax.set_title(majors[sp][m])
        ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
        ax.set_yticks([0,100])
        ax.axhline(50, c=cb_light_grey, alpha=0.3)
    
        if plot_no == 1:
            ax.text(2002, 85,'Women' )
            ax.text(2005, 10,'Men' )
        elif plot_no == 2:
            ax.text(2002, 76, 'Women')
            ax.text(2005, 18, 'Men')
        elif plot_no == 3:
            ax.text(2002, 93, 'Women')
            ax.text(2005, 6, 'Men')
        elif plot_no == 14:
            ax.tick_params(labelbottom='on')
        elif plot_no == 16:
            ax.text(2002, 8, 'Women')
            ax.text(2003, 88, 'Men')
            ax.tick_params(labelbottom='on')
        elif plot_no == 18:
            ax.text(2002, 30, 'Women')
            ax.text(2003,65, 'Men')
            ax.tick_params(labelbottom='on')


            
#plt.show()
plt.savefig("gender_degrees.png")
```
