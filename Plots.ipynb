# Pyplot fun

```python
%matplotlib inline 

import pandas as pd
import matplotlib.pyplot as plt

recent_grads = pd.read_csv("recent-grads.csv")
print(recent_grads.iloc[0])
recent_grads.describe()
```

Drop rows with missing values.

```python
raw_data_count = len(recent_grads.index)
print(" Raw data row number = ",raw_data_count)
recent_grads = recent_grads.dropna()
cleaned_data_count = len(recent_grads.index)
print(" Data without NaN row number = ",cleaned_data_count)
```

### 1. Scatter plots

```python
recent_grads.plot(x='Sample_size', y='Employed', kind='scatter', title='Employed vs. Sample_size', figsize=(5,5))
```

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Full_time vs. Median', figsize=(5,5))
```

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Sample_size vs. Unemployment_rate', figsize=(5,5))
```

Is there any link between the number of full-time employees and median salary?

NO

```python
recent_grads.plot(x='ShareWomen', y='Median', kind='scatter', title='ShareWomen vs. Median', figsize=(5,5))
```

<p>Do students that majored in subjects that were majority female make more money?</p>

It looks like no.


```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', title='ShareWomen vs. Unemployment_rate', figsize=(5,5))
```

```python
recent_grads.plot(x='Men', y='Median', kind='scatter', title='Men vs. Median', figsize=(5,5))
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter', title='Women vs. Median', figsize=(5,5))
```

### 3. Scatter matrix plot

```python
from pandas.tools.plotting import scatter_matrix

scatter_matrix(recent_grads[['Sample_size', 'Median']], figsize=(10,10))
```

```python
scatter_matrix(recent_grads[['Sample_size', 'Median','Unemployment_rate']], figsize=(10,10))
```

### 4. Bar plots

```python
recent_grads[:10].plot.bar(x='Major', y='Women')
```

```python
#lowes paying majors

recent_grads.sort_values('Median')[:10].plot.bar(x='Major', y='ShareWomen')
```

```python
#highest paying majors

recent_grads.sort_values('Median', ascending=False)[:10].plot.bar(x='Major', y='ShareWomen')
```

So sad... 10 of the lowest paying jobs are done mostly by women
